```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#LCOH

data = [[8.34, 8.34, 8.71, 8.71],
        [8.80, 8.80, 9.19, 9.19],
        [8.96, 8.92, 9.28, 9.22],
        [9.75, 9.70, 10.08, 10.00]]

X = np.arange(4)
fig = plt.figure(figsize=(8, 5))
ax = fig.add_axes([0,0,1,1])

# define the width of each bar and the spacing between them
bar_width = 0.15
bar_spacing = 0.05

ax.bar(X, data[0], color='b', width=bar_width, label='Scenario 1')
ax.bar(X + bar_width + bar_spacing, data[1], color='g', width=bar_width, label='Scenario 2')
ax.bar(X + 2 * (bar_width + bar_spacing), data[2], color='r', width=bar_width, label='Scenario 3')
ax.bar(X + 3 * (bar_width + bar_spacing), data[3], color='orange', width=bar_width, label='Scenario 4')

# set the x-axis limits
ax.set_xlim(-bar_width, len(X) - 1 + 4 * (bar_width + bar_spacing))

# set the x-axis tick marks in the middle of each group of bars
ax.set_xticks(X + 1.5 * (bar_width + bar_spacing))
ax.set_xticklabels(['Case 1a \n 70% (H2) / 30% (E) \n WC6 - 100 km', 'Case 1b \n 30% (H2) / 70% (E) \n WC6 - 100 km', 'Case 2a \n 70% (H2) / 30% (E) \n WC12 - 180 km', 'Case 2b \n 30% (H2) / 70% (E) \n WC12 - 180 km'])

# set y-axis limits
ax.set_ylim(0, 14)

# add legend
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# define the font size for the values
value_fontsize = 8

# loop over the data and add the values above each bar
for i in range(len(X)):
    for j in range(len(data)):
        ax.text(X[i] + j * (bar_width + bar_spacing), data[j][i]+0.2, f'{data[j][i]:.2f}', ha='center', fontsize=value_fontsize)

# set the title
ax.set_title('Levelised Cost of Hydrogen [EUR / kg]', fontsize=12)

plt.show()
```

```python
data = [[27.9, 25.6, 55.7, 51.1],
        [2.1, 8.2, 1.7, 13.4],
        [23.4, 21.5, 46.8, 43.0],
        [1.2, 6.2, 0.29, 10.1],
        [28.7, 26.5, 56.8, 52.3],
        [0.4, 7.0, -1.1, 11.7],
        [25.3, 23.6, 50.0, 46.5],
        [-1.4, 3.8, -4.3, 6.0]]

X = np.arange(4)
fig = plt.figure(figsize=(10, 6))
ax = fig.add_axes([0,0,1,1])

# define the width of each bar and the spacing between them
bar_width = 0.1
bar_spacing = 0.01

ax.bar(X, data[0], color='b', width=bar_width, label='Scenario 1 - Total discounted costs')
ax.bar(X + bar_width + bar_spacing, data[1], color='white', edgecolor='b', width=bar_width, hatch='\\', label='Scenario 1 - Final year cum. NPV (with revenues)')
ax.bar(X + 2 * (bar_width + bar_spacing), data[2], color='r', width=bar_width, label='Scenario 2 - Total discounted costs')
ax.bar(X + 3 * (bar_width + bar_spacing), data[3], color='white', edgecolor='r', width=bar_width, hatch='\\', label='Scenario 2 - Final year cum. NPV (with revenues)')
ax.bar(X + 4 * (bar_width + bar_spacing), data[4], color='g', width=bar_width, label='Scenario 3 - Total discounted costs')
ax.bar(X + 5 * (bar_width + bar_spacing), data[5], color='white', edgecolor='g', width=bar_width, hatch='\\', label='Scenario 3 - Final year cum. NPV (with revenues)')
ax.bar(X + 6 * (bar_width + bar_spacing), data[6], color='orange', width=bar_width, label='Scenario 4 - Total discounted costs')
ax.bar(X + 7 * (bar_width + bar_spacing), data[7], color='white', edgecolor='orange', width=bar_width, hatch='\\', label='Scenario 4 - Final year cum. NPV (with revenues)')

# set the x-axis limits
ax.set_xlim(-bar_width, len(X) - 1 + 8 * (bar_width + bar_spacing))

# set the x-axis tick marks in the middle of each group of bars
ax.set_xticks(X + 3.5 * (bar_width + bar_spacing))
ax.set_xticklabels(['Case 1a \n 70% (H2) / 30% (E) \n WC6 - 100 km', 'Case 1b \n 30% (H2) / 70% (E) \n WC6 - 100 km', 'Case 2a \n 70% (H2) / 30% (E) \n WC12 - 180 km', 'Case 2b \n 30% (H2) / 70% (E) \n WC12 - 180 km'])

# set y-axis limits
ax.set_ylim(-6, 60)

# add horizontal line at y=0
ax.axhline(0, color='black')

# add legend
ax.legend(bbox_to_anchor=(1.01, 1), loc='upper left', fontsize = 8)

# define the font size for the values
value_fontsize = 8

# loop over the data and add the values above each bar
for i in range(len(X)):
    for j in range(len(data)):
        if data[j][i] < 0:
            ax.text(X[i] + j * (bar_width + bar_spacing), data[j][i]-1.5, f'{data[j][i]:.2f}', ha='center', fontsize=value_fontsize)
        else:
            ax.text(X[i] + j * (bar_width + bar_spacing), data[j][i]+0.5, f'{data[j][i]:.2f}', ha='center', fontsize=value_fontsize)

# set the title
ax.set_title('Total costs and cumulative NPV in final year [billion EUR]', fontsize=15)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [[12.8, 0.40, 2.20, 7.00, 8.20],
        [22.4, -1.10, 1.70, 11.7, 13.4]]

X = np.arange(5)
fig = plt.figure(figsize=(10, 6))
ax = fig.add_axes([0,0,1,1])

# define the width of each bar and the spacing between them
bar_width = 0.3
bar_spacing = 0.10

# assign colors to scenarios
scenario_colors = ['tab:blue', 'b', 'g', 'b', 'g']

# assign patterns to cases
case_patterns = ['..', '\\\\']

for i in range(2):
    ax.bar(X + i * (bar_width + bar_spacing), data[i], color='w', width=bar_width, edgecolor='black',
           hatch=case_patterns[i], label=f'Case {i+1} ({i*6+6} GW, {i*80+100} km offshore)',
           alpha=0.8, zorder=3)

# set the colors of the scenarios
for i in range(len(scenario_colors)):
    ax.patches[i*5].set_facecolor(scenario_colors[i])
    ax.patches[i*5+1].set_facecolor(scenario_colors[i])

# set the x-axis limits
ax.set_xlim(-bar_width, len(X) - 1 + (bar_width + bar_spacing))

# set the x-axis tick marks in the middle of each group of bars
ax.set_xticks(X + 0.5 * (bar_width + bar_spacing))
ax.set_xticklabels(['Reference',
                    'Scenario 1 \n Onshore electrolysis \n 70% (H2) / 30% (E)',
                    'Scenario 3 \n Offshore electrolysis \n 70% (H2) / 30% (E)',
                    'Scenario 1 \n Onshore electrolysis \n 30% (H2) / 70% (E)',
                    'Scenario 3 \n Offshore electrolysis \n 30% (H2) / 70% (E)'], fontsize=11)

# set y-axis limits
ax.set_ylim(-5, 25)

# add horizontal line at y=0
ax.axhline(0, color='black')

# add legend
ax.legend(bbox_to_anchor=(1.01, 1), loc='upper left', fontsize=11)

# define the font size for the values
value_fontsize = 11

# loop over the data and add the values above each bar
for i in range(len(X)):
    for j in range(len(data)):
        if data[j][i] < 0:
            ax.text(X[i] + j * (bar_width + bar_spacing), data[j][i]-1.5, f'{data[j][i]:.2f}', ha='center', fontsize=value_fontsize)
        else:
            ax.text(X[i] + j * (bar_width + bar_spacing), data[j][i]+0.5, f'{data[j][i]:.2f}', ha='center', fontsize=value_fontsize)

                    
# set the title
ax.set_title('Cumulative NPV in final year [billion EUR]', fontsize=17)

plt.show()
```
