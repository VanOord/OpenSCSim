```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#LCOH

data = [[8.34, 8.34, 8.71, 8.71],
        [8.80, 8.80, 9.19, 9.19],
        [8.96, 8.92, 9.28, 9.22],
        [9.75, 9.70, 10.08, 10.00]]

X = np.arange(4)
fig = plt.figure(figsize=(8, 5))
ax = fig.add_axes([0,0,1,1])

# define the width of each bar and the spacing between them
bar_width = 0.15
bar_spacing = 0.05

ax.bar(X, data[0], color='b', width=bar_width, label='Scenario 1')
ax.bar(X + bar_width + bar_spacing, data[1], color='g', width=bar_width, label='Scenario 2')
ax.bar(X + 2 * (bar_width + bar_spacing), data[2], color='r', width=bar_width, label='Scenario 3')
ax.bar(X + 3 * (bar_width + bar_spacing), data[3], color='orange', width=bar_width, label='Scenario 4')

# set the x-axis limits
ax.set_xlim(-bar_width, len(X) - 1 + 4 * (bar_width + bar_spacing))

# set the x-axis tick marks in the middle of each group of bars
ax.set_xticks(X + 1.5 * (bar_width + bar_spacing))
ax.set_xticklabels(['Case 1a \n 70% (H2) / 30% (E) \n WC6 - 100 km', 'Case 1b \n 30% (H2) / 70% (E) \n WC6 - 100 km', 'Case 2a \n 70% (H2) / 30% (E) \n WC12 - 180 km', 'Case 2b \n 30% (H2) / 70% (E) \n WC12 - 180 km'])

# set y-axis limits
ax.set_ylim(0, 14)

# add legend
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# define the font size for the values
value_fontsize = 8

# loop over the data and add the values above each bar
for i in range(len(X)):
    for j in range(len(data)):
        ax.text(X[i] + j * (bar_width + bar_spacing), data[j][i]+0.2, f'{data[j][i]:.2f}', ha='center', fontsize=value_fontsize)

# set the title
ax.set_title('Levelised Cost of Hydrogen [EUR / kg]', fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for each case
cases = ['All components (10%)', 'Foundation & Cables (30%)', 'Foundation & Cables (5%)',
         'Turbines (30%)', 'Turbines (5%)', 'Electrolysis system (30%)',
         'Electrolysis system (5%)', 'Artificial island (30%)', 'Artificial island (5%)']
lcoh_means = [9.20, 9.23, 9.22, 9.23, 9.23, 9.22, 9.22, 9.22, 9.23]
lcoh_stdevs = [0.39, 0.72, 0.36, 0.87, 0.33, 0.59, 0.37, 0.42, 0.41]

# Create the plot
fig, ax = plt.subplots(figsize=(15,6), facecolor='white')
ax.errorbar(cases, lcoh_means, yerr=lcoh_stdevs, fmt='o', markersize=10, color='red', capsize=5, ecolor='black')

# Add axis labels and title
plt.xlabel('Component (standard deviation)', fontsize=14)
plt.ylabel('LCOH (/kg)', fontsize=14)
plt.title('Analysing the sensitivity of LCOH to component price uncertainty', fontsize = 18)

# Add a horizontal line for LCOH mean
plt.axhline(y=9.20, color='black', linestyle='--')

plt.xticks(rotation=30)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for each case
cases = ['Foundation & Cables (30%)', 'Foundation & Cables (5%)',
         'Turbines (30%)', 'Turbines (5%)', 'Electrolysis system (30%)',
         'Electrolysis system (5%)', 'Artificial island (30%)', 'Artificial island (5%)']
lcoh_stdevs = [0.72, 0.36, 0.87, 0.33, 0.59, 0.37, 0.42, 0.41]
lcoh_base = 0.39  # Base case is the first case
lcoh_pcts = [(x-lcoh_base)/lcoh_base * 100 for x in lcoh_stdevs]  # Percentage change from base case

# Define the colors for the bars
colors = ['#1f77b4']*2 + ['#ff7f0e']*2 + ['#2ca02c']*2 + ['#d62728']*2

# Create the bar chart
fig, ax = plt.subplots(figsize=(15,6))
bars = ax.bar(cases, lcoh_pcts, color=colors, capsize=10)

# set y-axis limits
ax.set_ylim(-30, 140)

# Add a horizontal line at 0%
plt.axhline(y=0, color='gray', linestyle='--', linewidth=1)

# Add axis labels and title
plt.xlabel('Component (standard deviation)', fontsize=14)
plt.ylabel('Change in standard deviation (%)', fontsize=14)
plt.title('Change in standard deviation of LCOH due to varying component price uncertainty', fontsize = 18)

plt.xticks(rotation=30)

# Add percentages on top of the bars, below for negative values
for bar in bars:
    height = bar.get_height()
    if height >= 0:
        ax.annotate(f'{height:.0f}%', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points", ha='center', va='bottom', fontsize=12)
    else:
        ax.annotate(f'{height:.0f}%', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, -3),  # 3 points vertical offset
                    textcoords="offset points", ha='center', va='top', fontsize=12)

# Show the plot
plt.show()
```
