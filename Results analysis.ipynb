```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Case 1.1

data = [[8.34, 8.80, 9.02, 9.81],
        [0.33, 0.35, 0.15, 0.19],
        [2.12, 1.18, 0.4, -1.41],]

X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# define the width of each bar and the spacing between them
bar_width = 0.2
bar_spacing = 0.1

ax.bar(X, data[0], color='b', width=bar_width, label='LCOH [EUR / kg]')
ax.bar(X + bar_width + bar_spacing, data[1], color='g', width=bar_width, label='LCOE [EUR / kWh]')
ax.bar(X + 2 * (bar_width + bar_spacing), data[2], color='r', width=bar_width, label='Final NPV [billion EUR]')

# set the x-axis labels
ax.set_xticks(X + bar_width + bar_spacing)
ax.set_xticklabels(['Scenario 1', 'Scenario 2', 'Scenario 3', 'Scenario 4'])

# set y-axis limits
ax.set_ylim(-2, 11)

# add legend
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# add horizontal line at y=0
ax.axhline(0, color='k')

# set the title
ax.set_title('Case 1.1 \n Wind park 100 km from the coast \n W2OH (70%) - W2E (30%) - PD30 - WC6 / C(D, P, IC, R) - O(Y, I, D) - E - DCF', fontsize=12)

plt.show()
```
